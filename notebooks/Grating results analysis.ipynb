```python
import sys
sys.path.append('..')

import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter, MultipleLocator
import pandas as pd
import numpy as np

import brainbox.physiology as phys
```

```python
# Define your model here
def model ():
    pass

gratingsProber = phys.GratingsProber(model, amplitude=0.5, rf_w=20, rf_h=20, n_timesteps=10, dt=1)

# Or preload precomputed example results
gratingsProber.grating_results = pd.read_pickle('./misc/grating_results.pickle')
```

```python
# Plot tuning curve
theta, tuning_curve = gratingsProber.orientation_tuning_curve
plt.polar(theta, tuning_curve)
plt.title('Direction tuning curve')
plt.gca().xaxis.set_major_formatter(FuncFormatter(
   lambda val,pos: '{:.3g}$\pi$'.format(val/np.pi) if val !=0 else '0'
))
plt.gca().xaxis.set_minor_locator(MultipleLocator(base=np.pi))
plt.show()

# Print grating results
print('Preferred direction:', gratingsProber.preferred_direction)
print('Preferred orientation:', gratingsProber.preferred_orientation)
print('Preferred spatial frequency:', gratingsProber.preferred_spatial_freq)
print('Preferred temporal frequency:', gratingsProber.preferred_temporal_freq)
print('Preferred output:', gratingsProber.preferred_model_output)
print('Orientation selectivity index:', gratingsProber.orientation_selectivity_index)
print('Direction selectivity index:', gratingsProber.direction_selectivity_index)
print('Circular variance:', gratingsProber.circular_variance)
print('Orientation bandwidth:', gratingsProber.orientation_bandwidth)
```
